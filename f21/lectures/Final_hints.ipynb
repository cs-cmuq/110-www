# Nice looking bar charts

```python
# this code just creates some random data!

import random

import math

age_range = [18, 90]
age_intervals = 6
one_age_interval = math.ceil( (age_range[1] - age_range[0] + 1) / age_intervals)

# this generates lists (of length age_intervals) with random integer numbers 
counts_by_age = random.sample( range(1, 100), age_intervals)
counts_by_belt =  random.sample( range(1, 50), age_intervals)

labels = []

start_age = age_range[0]

for i in range(age_intervals):
    end_age = start_age + one_age_interval -1
    labels.append( str(start_age) + '-' + str(end_age) ) #str( age_range[0] + (i * age_interval) ))
    start_age = end_age + 1
```

```python
labels
```

```python
import matplotlib.pyplot as plt  # import the module

import matplotlib as mpl
mpl.rcParams['figure.dpi']= 130  # set the resolution to x dpi
```

```python
plt.title("Bar Chart")


plt.bar(range(0, age_intervals),        
        counts_by_age,
        color='b',
        align='edge',
        width = 0.5,
        tick_label = labels)

plt.bar(range(0, age_intervals),        
        counts_by_belt,
        align='center',
        width = 0.3,
        color='r')

plt.show()
```

# Make multiple plots side by side and/or over multiple rows

```python
import matplotlib.pyplot as plt  # import the module

import matplotlib as mpl
mpl.rcParams['figure.dpi']= 130  # set the resolution to x dpi

# this defines a plot which is composed of one row of three plots
figure, subplots = plt.subplots(1, 3, figsize=(12, 3))

x = [1,5,3,9,0,6,10,4,12]
y = []
for i in x:
    y.append(i**3)
    
subplots[0].scatter(x, y, s=1)
subplots[0].set_title('Plot 1: some cubic transformation')

subplots[1].plot(range(3,60,3), c='r')
subplots[1].set_title('Plot 2: a line')

subplots[2].bar(['Male', 'Female'], [100, 120])
subplots[2].set_title('Plot 3: a simple bar plot')

plt.subplots_adjust(wspace=0.3)
plt.show()
```

This makes a *matrix* of plots

```python
import matplotlib.pyplot as plt  # import the module

import matplotlib as mpl
mpl.rcParams['figure.dpi']= 130  # set the resolution to x dpi

# this defines a plot which is composed of one row of two plots
figure, subplots = plt.subplots(2, 2, figsize=(10, 8))

subplots[0,0].scatter(list(range(1,100,2)), list(range(0, 50, 1)), s=1)
subplots[0,0].set_title('Plot 1')

subplots[0,1].plot(range(3,60,3), c='r')
subplots[0,1].set_title('Plot 2')

subplots[1,0].scatter(list(range(1,100,2)), list(range(0, 50, 1)), s=1)
subplots[1,0].set_title('Plot 3')

subplots[1,1].plot(range(3,60,3), c='r')
subplots[1,1].set_title('Plot 4')

plt.subplots_adjust(wspace=0.3, hspace=0.35)
plt.show()
```

```python
plt.figure(figsize=(9,7))

y1 = range(-1, 10, 1)
y2 = range(2, 40, 5)

legends = [0]*2

legends[0], = plt.plot(y1, label='plot 1')
legends[1], = plt.plot(y2, label='plot 2')

plt.legend(handles=legends)
plt.show()
```

# Get multiple legends for one single plot

```python
# just some example data and labels
x = range(1,10)
y = range(6, 15)
data_labels = [0,1,0,0,1,1,1,0,0]

scatterplot_data = plt.scatter( x, y, c=data_labels)

lines, lab = scatterplot_data.legend_elements()

legend_labels = ['a', 'b']
plt.legend(lines, legend_labels)

plt.show()
```
