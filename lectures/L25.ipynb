***
Recap on methods from the `random` module:

- `seed(seed_val = Null)` initializes the state of the random generator
- `uniform(from, to)` generates random float numbers uniformly in the *real* range `[from, to]`
- `randint(from, to)` generates random integer numbers uniformly in the *integer* range `{from, ..., to}`
- `choice(seq)` generates elements drawn uniformly from the *sequence* `seq`
***

```python
import random

new_seed = 76438
random.seed(new_seed)

v_min = 2.0
v_max = 5.5
n = 50
print('\n{} real numbers uniformly distributed between {} and {}:'.format(n, v_min, v_max))
for i in range(n):
    print('{:.2f} '.format(random.uniform(v_min, v_max)), end='')
    
v_min = 2
v_max = 5
n = 50
print('\n\n{} integer numbers uniformly distributed between {} and {}:'.format(n, v_min, v_max))
for i in range(n):
    print('{:d} '.format(random.randint(v_min, v_max)), end='')
    
v_min = 2
v = [0, 'to', 'be', 'or', 'not', 'to', 'be', 'that', 'is', 'the', 'question', 1]
n = 50
print('\n\n{} elements uniformly drawn from the sequence {}:'.format(n, v))
for i in range(n):
    print('{} '.format(random.choice(v)), end='')

```

Let's check with some detail how to generate data according to a *Gaussian distribution.* 
***

***
The method `gauss(mean, std_deviation)` returns data distributed according to a **Gaussian** (also termed a *Normal* distribution):

$$\mathcal{N}(x; \mu, \sigma) = \frac{1}{{\sigma \sqrt {2\pi}}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$


A Gaussian distribution is a identified by two parameters, the **mean** $\mu$, and the **standard deviation** $\sigma$. With a probability equal to about 0.68, a data value sampled according to a Gaussian distribution has a value in the range $[\mu -\sigma, \mu + \sigma]$. The probability for being close to the mean value $\mu$ is higher than the values in the other regions of the the range, with the probability monotonically decreasing from both sides according to the Gauss curve. Looking at the figure, it is clear that sampling a value outside of the interval $[\mu -2\sigma, \mu + 2\sigma]$ is very unlikely, it is expected to happen about in only the 5% of the cases.

 <img src="gaussian-labeled.png" width="700">

***

```python
gauss_distr = []
generations = 5000
mean = 0
std_dev = 1.5
for i in range(generations):
    gauss_distr.append(random.gauss(mean, std_dev))

plt.figure(figsize = (7,5))
plt.title('Normally distributed random data')
plt.xlabel('Trial')
ylabel = '$G({}, {})$'.format(mean, std_dev)
plt.ylabel(ylabel)

rgb = (random.uniform(0,1), random.uniform(0,1), random.uniform(0,1))
plt.plot(gauss_distr, linestyle='None', marker='.', markersize=1, color=rgb)
plt.show()
```

***
In order to plot the expected Gaussian function $G(\mu, \sigma)$, we need to generate the pairs $(x, G(x; \mu, \sigma))$. The function `gen_gaussian(mean, variance, xmin, xmax, delta_x)` generates the desired pairs pairs in a given interval for the $x$ variable. 

Note the use of the `arange` method from `numpy` to generate a range of floats. `numpy` has also been used as a replacement for a few functions provided by `math`
***

```python
import numpy as np
def gen_gaussian(mean, std_dev, xmin, xmax, dx):
    g = [[], []]
    for x in np.arange(xmin, xmax, dx):
        g[0].append(x)
        g[1].append( (1/(np.sqrt(2 * np.pi) * std_dev)) * np.exp(-((x - mean)**2) / (2 * std_dev) ))
    return(g)
```

```python
mean = 2
std_dev = 2
xmin = 2 - 2.5*std_dev
xmax = 2 + 2.5*std_dev
delta_x = 0.1
gaussian_values = gen_gaussian(mean, std_dev, xmin, xmax, delta_x)
plt.figure()
plt.xlabel('x')
plt.ylabel('$G({}, {})$'.format(mean, std_dev))
plt.plot(gaussian_values[0], gaussian_values[1])
```

```python
# The MC model of the Buffon experiment, return the estimate of pi, 
# as well as the list of the sampled points (for later)
#
def pi_estimate_by_monte_carlo_trial(num_points=1000):
    pts_x = []
    pts_y = []
    in_circle = 0
    for p in range(num_points):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if np.sqrt(x*x + y*y) <= 1:
            in_circle += 1
        pts_x.append(x)
        pts_y.append(y)
    return (4 * (in_circle / num_points), pts_x, pts_y)
```

```python
# Average all MC measures from the trial 
#
def pi_estimate(num_trials, num_points):
    estimates = []    
    for t in range(num_trials):
        # note: pts_x and pts_y are not used here
        pi_estimate, pts_x, pts_y = pi_estimate_by_monte_carlo_trial(num_points)
        estimates.append(pi_estimate)
    mean_estimates = sum(estimates) / num_trials
    stddev_estimates = std_dev(estimates)
    
    return (mean_estimates, stddev_estimates, estimates)
```

***
Given a set $V$ of $n$ measures, $V = \{v_1, v_2, \ldots, v_n\}$, the *sample mean* $\mu$ is computed as:

$$\bar{\mu} = \frac{1}{n}\sum_{i=1}^n v_i .$$

The sample mean should be regarded as an estimate of the the *true mean* $\mu$ of the population from where the measures where drawn from. In the limit $n \to \infty$ the sample mean converges to true mean, $\bar{\mu} \to \mu$.

The dispersion of a set of measures about their mean, is quantified by the standard deviation $\sigma$, which is the average of the sum of the quadratic differences from the mean. The sample standard deviation $\bar{\sigma}$ can be computed as follows:

$$\bar{\sigma} = \sqrt{\frac{1}{(n-1)}\sum_{i=1}^n (v_i - \bar\mu})^2 .$$
***

```python
def std_dev(values):
    mean = sum(values) / len(values)
    std = 0.0
    for v in values:
        std += (v - mean) ** 2
    return np.sqrt(std / (len(values)-1))

v = [2,2,2,2,2,]
print('Standard deviation: {:.3f}'.format(std_dev(v)))

v = [1,2,3,4,5]
print('Standard deviation: {:.3f}'.format(std_dev(v)))

v = [1,2,3,4,10]
print('Standard deviation: {:.3f}'.format(std_dev(v)))
```

```python
def pi_estimate_with_given_precision(precision, num_trials = 50, num_points = 100):
    all_mc_estimates = []
    stddev_estimate = precision
    while stddev_estimate  >= precision:
        # note: mc_estimates is not used here
        estimate, stddev_estimate, mc_estimates = pi_estimate(num_trials, num_points)
        all_mc_estimates.append(mc_estimates)
        num_points *= 2
        
    print('Exact \u03C0 value: 3.141592653')
    print('Estimate of \u03C0: {:10.8f} (estimated error: {:3.4f}, #points: {})'.format(estimate, 
                                                                                    stddev_estimate,
                                                                                    num_points))
    return all_mc_estimates

v = pi_estimate_with_given_precision(0.01)
```

Let's start by drawing the circle by computing the points, which is relatively easy. At this aim, we can exploit the *parametric equations* of a circle, that say that the $(x,y)$ coordinates of the points of the unit circle centered in $(0,0)$ are obtained as follows:

$$ \begin{cases} x = cos(t) \\[1ex] y = sin(t) \end{cases} \quad\quad \text{where}\quad t\in [0, 2\pi]$$.

Once we have the circle, we can plot the points generated during the Monte Carlo trial.
***

```python
x = []
y = []
dt = 0.1
for t in np.arange(0, 2*np.pi + dt, dt):
    x.append(np.cos(t))
    y.append(np.sin(t))

plt.figure()
plt.figure(figsize = (7,7))
plt.title('Monte Carlo simulation of a Buffon experiment')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xticks(np.arange(-1,2,1))
plt.yticks(np.arange(-1,2,1))
plt.plot(x, y)

estimate, pts_x, pts_y = pi_estimate_by_monte_carlo_trial(5000)
plt.scatter(pts_x, pts_y, s=2, color='r')
plt.show()

print('Estimate of \u03C0 from Monte Carlo trial: {:.3f}'.format(estimate))
```

**Boxplots** summarize a lot of relevent statistical information about data. The box extends from the lower to upper *quartile values* of the data, with a line at the *median*. The whiskers extend from the box to show the *range* of the data. *Flier points* are those past the end of the whiskers.

A *quartile* is a statistical term describing a division of data into *four defined intervals* based upon the values of the data. The quartile breaks down the data into quarters so that 25% of the data have value less than the value of the lower quartile (Q1), 50% are less than the median quartile (Q2), and 75% are less than the upper quartile (Q3).
