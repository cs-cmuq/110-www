## 3.2 Display Plots With Python

```python
import matplotlib.pyplot as plt
import random

y = random.sample(range(10, 100), 50)

plt.figure(figsize = (7,5))
plt.title('A random curve!')
          
plt.plot(y, color='r')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x), 'c--', linewidth=3)
plt.show()
```

### 3.4.2 Clearing Up History

Clear it up from (Kernel) drop down menu. Press Kernel >> Run and Clear Output. After this operation, everything in the history is gone.

A notebook kernel is a “computational engine” that executes the code contained in a notebook document. The kernel performs the computation and produces the results
